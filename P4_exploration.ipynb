## Bivariate analysis

```python
# Very strong to strong correlations
featDict = {'Electricity(kWh)': 'Electricity(kBtu)',
            'NaturalGas(therms)': 'NaturalGas(kBtu)',
            'SiteEUI(kBtu/sf)': 'SiteEUIWN(kBtu/sf)',
            'SourceEUI(kBtu/sf)': 'SourceEUIWN(kBtu/sf)',
            'SiteEnergyUse(kBtu)': ['SiteEnergyUseWN(kBtu)', 'PropertyGFATotal', 'PropertyGFABuilding(s)', 'LargestPropertyUseTypeGFA', 'NumberofBuildings'],
            'GHGEmissions(MetricTonsCO2e)': ['SiteEnergyUse(kBtu)', 'PropertyGFATotal', 'PropertyGFABuilding(s)', 'LargestPropertyUseTypeGFA', 'ENERGYSTARScore'],
            'SiteEUIWN(kBtu/sf)': ['GHGEmissionsIntensity(kgCO2e/ft2)', 'ENERGYSTARScore']
            }

# Do not consider NA values for correlation computation
naList = list(featDict.keys())
for val in featDict.values():
    if isinstance(val, list):
        for nested_val in val:
            naList.append(nested_val)
    else:
        naList.append(val)

nonNaResData = nonResDataWtOut.copy()
print(nonNaResData.shape)
nonNaResData = nonResDataWtOut.dropna(subset=naList)
print(nonNaResData.shape)
print("\n")

import scipy.stats as st

# Bivariate analysis of numerical features
for feat1, feat2 in featDict.items():
    if isinstance(feat2, list):
        for el in feat2:
            print(feat1+" versus "+el+"\n")
            # Compute correlation coefficient 
            pears_coeff, pVal = st.pearsonr(nonNaResData[feat1], nonNaResData[el])
            print('P-Value: {}'.format(pVal))
            print('Pearson: {}'.format(pears_coeff))
            print('Cov: {}'.format(np.cov(nonNaResData[feat1],nonNaResData[el],ddof=0)[1,0]))

            # Scatter plot
            plt.plot(nonNaResData[feat1], nonNaResData[el],'o',alpha=0.5)
            plt.xlabel(feat1)
            plt.ylabel(el)
            plt.show()
    else:
        print(feat1+" versus "+feat2+"\n")
    
        # Compute correlation coefficient 
        pears_coeff, pVal = st.pearsonr(nonNaResData[feat1], nonNaResData[feat2])
        print('P-Value: {}'.format(pVal))
        print('Pearson: {}'.format(pears_coeff))
        print('Cov: {}'.format(np.cov(nonNaResData[feat1],nonNaResData[feat2],ddof=0)[1,0]))

        # Scatter plot
        plt.plot(nonNaResData[feat1], nonNaResData[feat2],'o',alpha=0.5)
        plt.xlabel(feat1)
        plt.ylabel(feat2)
        plt.show()
```

## Create a new  categorical feature

```python
# Create a new colomn

print("Min YearBuilt: ", nonResDataWtOut['YearBuilt'].min())
print("Max YearBuilt: ", nonResDataWtOut['YearBuilt'].max())

print(nonResDataWtOut.shape)
nonResDataWtOut = nonResDataWtOut.assign(YearBuilt_period=lambda x: 'Between ' + ((x['YearBuilt']//10)*10).astype(str) + ' and ' + (((x['YearBuilt']//10)+1)*10).astype(str))
print(nonResDataWtOut.shape)
display(nonResDataWtOut.head())

print(nonResDataWtOut['YearBuilt'].nunique())
print(nonResDataWtOut['YearBuilt_period'].nunique())
print(nonResDataWtOut['YearBuilt_period'].unique())
```

```python
# Bivariate analysis of categorical features
categ_features = ['BuildingType', 'PrimaryPropertyType', 'Neighborhood', 'LargestPropertyUseType', 'YearBuilt_period']

def eta_squared(x,y):
    mean_y = y.mean()
    classes = []
    x_modalities = x.unique()
    x_modalities = x_modalities[pd.notna(x_modalities)]
    x_modalities.sort()
    for classe in x_modalities:
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'mean_classe': yi_classe.mean()})
    SCT = sum([(yj-mean_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['mean_classe']-mean_y)**2 for c in classes])
    return SCE/SCT


dep_vars = ['SiteEnergyUse(kBtu)', 'GHGEmissions(MetricTonsCO2e)']

for X in categ_features:
    modalities = nonResDataWtOut[X].unique()
    modalities = modalities[pd.notna(modalities)]
    modalities.sort()


    for Y in dep_vars:
        nonNaResData = nonResDataWtOut.copy()
        print(nonNaResData.shape)
        nonNaResData = nonResDataWtOut.dropna(subset=[X, Y])
        print(nonNaResData.shape)
        print("\n")

        groups = []
        for m in modalities:
            groups.append(nonNaResData[nonNaResData[X]==m][Y])

        print("Eta-squared: ", eta_squared(nonNaResData[X],nonNaResData[Y]))
        medianprops = {'color':"black"}
        meanprops = {'marker':'o', 'markeredgecolor':'black',
                    'markerfacecolor':'firebrick'}
        fig = plt.figure(figsize=(15,8))

        plt.xlabel(Y)
        plt.ylabel(X)
        plt.boxplot(groups, labels=modalities, showfliers=False, medianprops=medianprops, 
                    vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)
        plt.show()
```
