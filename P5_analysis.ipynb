```python
%matplotlib inline
sns.set()

plt.style.use('seaborn')
# Set appropriate colors map for color blindness 
plt.set_cmap('coolwarm')
ax = plt.axes()

# Univariate analysis of numerical features
num_features = ['payment_value', 'price', 'freight_value']

for feat in num_features:
    print(feat)
    print("mean:\n",merged_df[feat].mean())
    print("med:\n",merged_df[feat].median())
    print("std:\n",merged_df[feat].std(ddof=0))
    print("mod:\n",merged_df[feat].mode())
    merged_df[feat].hist(density=True)
    plt.title("Histogram")
    plt.xlabel(feat)
    plt.show()
    plt.title("Boxplot")
    merged_df.boxplot(column=feat, vert=False)
    plt.show()
```

```python
# Bivariate analysis of numerical features

import scipy.stats as st

num_features = ['payment_value', 'price', 'freight_value']

for ft1 in num_features:
    for ft2 in num_features:
        if ft1 != ft2:
            print(ft1 + " versus " + ft2 + "\n")
            pears_coeff, pVal = st.pearsonr(merged_df[ft1], merged_df[ft2])
            print('P-Value: {}'.format(pVal))
            print('Pearson: {}'.format(pears_coeff))
            print('Cov: {}'.format(np.cov(merged_df[ft1], merged_df[ft2], ddof=0)[1,0]))
            
            plt.plot(merged_df[ft1], merged_df[ft2],'o', alpha=0.5)
            plt.xlabel(ft1)
            plt.ylabel(ft2)
            plt.show()

# Check correlation between payment value and the sum of both price and freight_value
print("Payment value Versus price and freight_value sum: ")
pears_coeff, pVal = st.pearsonr(merged_df['payment_value'], merged_df['price'] + merged_df['freight_value'])
print('P-Value: {}'.format(pVal))
print('Pearson: {}'.format(pears_coeff))

plt.plot(merged_df['payment_value'], merged_df['price'] + merged_df['freight_value'],'o', alpha=0.5)
plt.xlabel('payment_value')
plt.ylabel('Sum of price and freight_value')
plt.show()
```

```python
from sklearn import preprocessing
from sklearn import decomposition

# Convert to int:
# order_id
# customer_id
# review_id
# customer_unique_id
# product_id
# seller_id

# One hot encoding:
# order_status
# payment_type

# Hashing:
# customer_state
# seller_state
# product_category_name


simplified_df = merged_df[['payment_sequential', 'payment_installments', 'payment_value', 'review_score',
                          'order_item_id', 'price', 'freight_value', 'product_photos_qty', 'order_id',
                          'customer_id', 'review_id', 'customer_unique_id', 'product_id',
                          'seller_id', 'payment_type', 'customer_state', 'seller_state',
                          'product_category_name']]

# Convert to number
id_col_list = ['order_id', 'customer_id', 'review_id', 'customer_unique_id', 'product_id', 'seller_id']
simplified_df_converted = simplified_df.copy()
for col in id_col_list:
    simplified_df_converted[col] = simplified_df[col].apply(lambda x: int(x, 16))

# One Hot Encoding
simplified_df_converted = pd.get_dummies(simplified_df_converted, columns=['payment_type'])

# Hashing
import hashlib
"""
hash_func:
From source dataframe to destination, apply a hash function based on sha256
"""
def hash_func(source_df, destination_df, *column):
    column_name = ''
    h = hashlib.sha256
    destination_df[column_name.join(column)] = pd.DataFrame(source_df[list(column)].values.sum(axis=1))[0].str.\
                                               encode('utf-8').apply(lambda x: (h(x).hexdigest().upper()))

# Apply hash function
hash_func(simplified_df_converted, simplified_df_converted, 'customer_state')
# Convert to int
simplified_df_converted['customer_state'] = simplified_df_converted['customer_state'].apply(lambda x: int(x, 16))

hash_func(simplified_df_converted, simplified_df_converted, 'seller_state')
simplified_df_converted['seller_state'] = simplified_df_converted['seller_state'].apply(lambda x: int(x, 16))

# Remove underscores
simplified_df_converted['product_category_name'] = simplified_df_converted['product_category_name'].\
                                                                            apply(lambda x: str(x).replace('_', ''))
# Apply hashing and convert to int
hash_func(simplified_df_converted, simplified_df_converted, 'product_category_name')
simplified_df_converted['product_category_name'] = simplified_df_converted['product_category_name'].apply(lambda x: int(x, 16))


display(simplified_df_converted.head())
print(simplified_df_converted.shape)
# Remove NA values
for col in simplified_df_converted.columns:
    if simplified_df_converted[col].isnull().mean()*100 > 1:
        print(col)
        print(simplified_df_converted[col].isnull().mean()*100)
simplified_df_converted = simplified_df_converted.dropna(how='any')
print(simplified_df_converted.shape)


X = simplified_df_converted.values

std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)

pca = decomposition.PCA(n_components=21) # 14, +6, +11(7+4) +3 = 34
pca.fit(X_scaled)

print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())


cumulative = np.cumsum(pca.explained_variance_ratio_)
print(cumulative)

plt.plot(range(0,21), cumulative, 'r')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.title('Explained Variance Ratio')
plt.show()

pcs = pca.components_

fig = plt.figure(figsize=(15,8))
for i, (x, y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    plt.plot([0, x], [0, y], color='k')
    plt.text(x, y, simplified_df_converted.columns[i], fontsize='14')

plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
```

# Feature engineering - Focus on customers

```python
display(merged_df.head())
print(merged_df.shape)
dups = merged_df[merged_df.duplicated(subset=['order_id'])]
display(dups.head())
print(dups.shape)
print(merged_df.duplicated(subset=['customer_unique_id']).sum())
print('Analyse rows with same order id:')
display(merged_df[(merged_df['order_id']==dups['order_id'].iloc[0])].head())
display(merged_df[(merged_df['order_id']==dups['order_id'].iloc[10])].head())

# First grouping
agg_df = merged_df.groupby(['order_id', 'product_id', 'customer_unique_id', 'payment_type', 'product_category_name']).\
                        agg({'review_score': 'mean', 'order_item_id': 'max', 'price': 'max', 'freight_value': 'max'})
agg_df = agg_df.reset_index()
print("Shape of first aggregation:")
print(agg_df.shape)
display(agg_df.head())


dups = agg_df[agg_df.duplicated(subset=['order_id'])]
display(dups.head())
print(dups.shape)
print('Analyse rows with same order id after first grouping:')
display(agg_df[(agg_df['order_id']==dups['order_id'].iloc[0])].head())

# One hot encoding of payment_type' and 'product_category_name'
agg_df_dummies = pd.get_dummies(agg_df, columns=['payment_type', 'product_category_name'])
print("Shape of dummies:", agg_df_dummies.shape)
agg_df_dummies.head()

print(agg_df_dummies.duplicated().sum())

dups = agg_df_dummies[agg_df_dummies.duplicated(subset=['order_id'])]
display(dups.head())
print(dups.shape)
print('Analyse rows with same order id after one hot encoding:')
display(agg_df_dummies[(agg_df_dummies['order_id']==dups['order_id'].iloc[0])].head())


agg_df_dummies_1 = agg_df_dummies.groupby(['order_id', 'product_id', 'customer_unique_id',
                                          'review_score', 'order_item_id', 'price', 'freight_value']).sum()
agg_df_dummies_1 = agg_df_dummies_1.reset_index()
print("Shape of dummies, after second grouping: ", agg_df_dummies_1.shape)
agg_df_dummies_1.head()

dups = agg_df_dummies_1[agg_df_dummies_1.duplicated(subset=['order_id'])]
display(dups.head())
print(dups.shape)
print('Analyse rows with same order id after second grouping:')
display(agg_df_dummies_1[(agg_df_dummies_1['order_id']==dups['order_id'].iloc[0])].head())

print(agg_df_dummies_1.duplicated(subset=['order_id', 'product_id', 'customer_unique_id']).sum())
print(agg_df_dummies_1.duplicated(subset=['order_id', 'product_id']).sum())
print(agg_df_dummies_1['customer_unique_id'].nunique())


# Create a new column: total_payment, order_item_id is equivalent to the quantity of the purchased product
agg_df_dummies_1['total_payment'] = agg_df_dummies_1['order_item_id']*(agg_df_dummies_1['price'] +
                                                                       agg_df_dummies_1['freight_value'])

display(agg_df_dummies_1[agg_df_dummies_1['order_item_id']==2].head())

col_list = list(agg_df_dummies_1.columns[7:])
print(len(col_list))

# Build dict
agg_dict = {'review_score': 'mean'}
agg_dict.update((idx, 'sum') for idx in col_list)
print(len(agg_dict))
agg_df_dummies_2 = agg_df_dummies_1.groupby('customer_unique_id').agg(agg_dict).reset_index()

print("Shape of dataframe after third grouping:", agg_df_dummies_2.shape)
display(agg_df_dummies_2.head())
```

# PCA on transformed dataframe

```python
from sklearn import preprocessing
from sklearn import decomposition

print(agg_df_dummies_2.shape)
transformed_df = agg_df_dummies_2.drop('customer_unique_id', axis=1)
print(transformed_df.shape)

X = transformed_df.values

std_scale = preprocessing.StandardScaler().fit(X)
X_scaled = std_scale.transform(X)

pca = decomposition.PCA(n_components=79)
pca.fit(X_scaled)

print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())


cumulative = np.cumsum(pca.explained_variance_ratio_)
print(cumulative)

plt.plot(range(0,79), cumulative, 'r')
plt.ylabel('Explained Variance')
plt.xlabel('Principal Components')
plt.title('Explained Variance Ratio')
plt.show()

pcs = pca.components_

fig = plt.figure(figsize=(15,8))
for i, (x, y) in enumerate(zip(pcs[0, :], pcs[1, :])):
    plt.plot([0, x], [0, y], color='k')
    plt.text(x, y, transformed_df.columns[i], fontsize='14')

plt.plot([-1, 1], [0, 0], color='grey', ls='--')
plt.plot([0, 0], [-1, 1], color='grey', ls='--')
plt.xlim([-1, 1])
plt.ylim([-1, 1])

```
